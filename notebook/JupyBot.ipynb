```python
from langchain.llms.base import LLM
from pydantic import BaseModel, Field
from typing import Optional, List
import google.generativeai as genai

class GeminiLLM(LLM, BaseModel):
    model_name: str = Field(default="models/gemini-1.5-flash")
    api_key: str = Field(...)

    class Config:
        extra = "allow"  # ✅ This allows unexpected fields like 'model' to be accepted

    def _call(self, prompt: str, stop: Optional[List[str]] = None) -> str:
        try:
            genai.configure(api_key=self.api_key)
            model = genai.GenerativeModel(model_name=self.model_name)
            response = model.generate_content(prompt)
            return response.text if hasattr(response, "text") else str(response)
        except Exception as e:
            raise ValueError(f"Gemini API error: {e}")

    @property
    def _llm_type(self) -> str:
        return "gemini"
```

```python
# Install required libraries
!pip install -q langchain google-generativeai faiss-cpu

from google.colab import userdata, drive
import os
import pickle
import google.generativeai as genai
from langchain.llms.base import LLM
from langchain.schema import Generation, LLMResult, Document
from langchain.vectorstores import FAISS
from langchain.chains import RetrievalQAWithSourcesChain
from langchain.embeddings.base import Embeddings





# Mount Google Drive
drive.mount('/content/drive')

# Load the vector index
file_path = "/content/drive/My Drive/vector_index.pkl"
vectorIndex = None
if os.path.exists(file_path):
    try:
        file_size = os.path.getsize(file_path)
        if file_size == 0:
            raise ValueError("The file is empty. Recreate the vector index.")
        with open(file_path, "rb") as f:
            vectorIndex = pickle.load(f)
        print(f"Vector index loaded successfully. Shape: {len(vectorIndex)} embeddings")
        if not vectorIndex or any(len(emb) == 0 for emb in vectorIndex):
            raise ValueError("Vector index contains empty or invalid embeddings.")
    except EOFError:
        print("Error: The file is empty or corrupted. Recreate the vector index.")
    except Exception as e:
        print(f"Error loading vector index: {e}")
else:
    raise FileNotFoundError(f"File {file_path} does not exist. Please run the embedding creation code first.")

# Define the original texts (must match the embeddings)
texts = ["Hello, how are you?", "I love coding!", "Google Colab is great!"]
if len(vectorIndex) != len(texts):
    raise ValueError(f"Number of embeddings ({len(vectorIndex)}) does not match number of texts ({len(texts)}). Recreate the vector index.")

# Custom Gemini Embeddings class
class GeminiEmbeddings(Embeddings):
    def __init__(self):
        self.api_key = os.getenv("GEMINI_API_KEY")
        genai.configure(api_key=self.api_key)

    def embed_documents(self, texts):
        embeddings = []
        for text in texts:
            try:
                result = genai.embed_content(
                    model="models/text-embedding-004",
                    content=text,
                    task_type="retrieval_query"
                )
                embeddings.append(result["embedding"])
            except Exception as e:
                print(f"Error embedding document '{text}': {e}")
                return None
        return embeddings

    def embed_query(self, text):
        try:
            result = genai.embed_content(
                model="models/text-embedding-004",
                content=text,
                task_type="retrieval_query"
            )
            return result["embedding"]
        except Exception as e:
            print(f"Error embedding query '{text}': {e}")
            return None

# Initialize the embeddings
embeddings = GeminiEmbeddings()

# Pair texts with their embeddings
text_embeddings = list(zip(texts, vectorIndex))

# Create the FAISS vector store
try:
    vectorstore = FAISS.from_embeddings(
        text_embeddings=text_embeddings,
        embedding=embeddings
    )
    print("FAISS vector store created successfully.")

    # Test the vector store
    test_query = "How are you?"
    results = vectorstore.similarity_search(test_query, k=1)
    print(f"Test search result: {results[0].page_content if results else 'No results'}")
except Exception as e:
    raise ValueError(f"Failed to create FAISS vector store: {e}")

class GeminiLLM(LLM):
    model_name: str = "gemini-1.5-flash"

    def __init__(self, api_key: str, **kwargs):
        super().__init__(**kwargs)
        genai.configure(api_key=api_key)
        self.model = genai.GenerativeModel(self.model_name)

    def _call(self, prompt: str, stop: list = None) -> str:
        try:
            response = self.model.generate_content(prompt)
            return response.text
        except Exception as e:
            raise ValueError(f"Gemini API error: {e}")

    @property
    def _llm_type(self) -> str:
        return "gemini"





# Create a retriever
retriever = vectorstore.as_retriever()

# Create the chain
chain = RetrievalQAWithSourcesChain.from_llm(
    llm=llm,
    retriever=retriever
)

```
